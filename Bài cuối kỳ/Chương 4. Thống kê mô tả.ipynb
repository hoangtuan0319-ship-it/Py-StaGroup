```python
# ======================================================
# ðŸ§  EDA - PhÃ¢n tÃ­ch thá»‘ng kÃª mÃ´ táº£ vÃ  trá»±c quan hÃ³a dá»¯ liá»‡u
# Dataset: Sleep_health_and_lifestyle_clean.csv
# ======================================================

import os
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Cáº¥u hÃ¬nh hiá»ƒn thá»‹
sns.set(style="whitegrid")
plt.rcParams['figure.dpi'] = 120
pd.set_option('display.max_columns', None)

# ======================================================
# 1ï¸âƒ£ Äá»c dá»¯ liá»‡u
# ======================================================
DATA_PATH = "/content/Sleep_health_and_lifestyle_clean.csv"   # Ä‘á»•i náº¿u cáº§n
df = pd.read_csv(DATA_PATH)

print("âœ… Dá»¯ liá»‡u Ä‘Ã£ náº¡p thÃ nh cÃ´ng!")
print(f"KÃ­ch thÆ°á»›c dá»¯ liá»‡u: {df.shape}")
display(df.head())

# ======================================================
# 2ï¸âƒ£ Kiá»ƒm tra kiá»ƒu dá»¯ liá»‡u vÃ  giÃ¡ trá»‹ thiáº¿u
# ======================================================
print("\nðŸ”¹ ThÃ´ng tin dá»¯ liá»‡u:")
df.info()

print("\nðŸ”¹ Sá»‘ lÆ°á»£ng giÃ¡ trá»‹ thiáº¿u:")
display(df.isna().sum())

# Biá»ƒu Ä‘á»“ giÃ¡ trá»‹ thiáº¿u
plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(), cbar=False)
plt.title("Biá»ƒu Ä‘á»“ phÃ¢n bá»‘ giÃ¡ trá»‹ thiáº¿u")
plt.show()

# ======================================================
# 3ï¸âƒ£ PhÃ¢n loáº¡i biáº¿n (numeric / categorical)
# ======================================================
numeric_cols = df.select_dtypes(include=np.number).columns.tolist()
categorical_cols = df.select_dtypes(exclude=np.number).columns.tolist()

print(f"\nðŸ”¹ Biáº¿n sá»‘ (numeric): {numeric_cols}")
print(f"ðŸ”¹ Biáº¿n phÃ¢n loáº¡i (categorical): {categorical_cols}")

# ======================================================
# 4ï¸âƒ£ Thá»‘ng kÃª mÃ´ táº£ cho biáº¿n sá»‘
# ======================================================
print("\nðŸ“Š Báº£ng thá»‘ng kÃª mÃ´ táº£ (describe):")
display(df[numeric_cols].describe().T)

# ThÃªm skewness vÃ  kurtosis
stats_df = pd.DataFrame({
    'Skewness': df[numeric_cols].skew(),
    'Kurtosis': df[numeric_cols].kurt()
})
print("\nðŸ”¹ Skewness & Kurtosis:")
display(stats_df)

# ======================================================
# 5ï¸âƒ£ PhÃ¡t hiá»‡n giÃ¡ trá»‹ ngoáº¡i lai (outliers) theo IQR
# ======================================================
outlier_summary = {}
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower) | (df[col] > upper)][col]
    outlier_summary[col] = len(outliers)

outlier_df = pd.DataFrame.from_dict(outlier_summary, orient='index', columns=['Outlier_count'])
print("\nðŸ”¹ Sá»‘ lÆ°á»£ng ngoáº¡i lai theo IQR:")
display(outlier_df)

# ======================================================
# 6ï¸âƒ£ Trá»±c quan hÃ³a phÃ¢n bá»‘ cá»§a biáº¿n sá»‘
# ======================================================
output_dir = Path("/content/eda_outputs")
output_dir.mkdir(exist_ok=True)

for col in numeric_cols:
    fig, axes = plt.subplots(1, 3, figsize=(15, 4))
    sns.histplot(df[col], kde=True, ax=axes[0])
    axes[0].set_title(f'Histogram: {col}')

    sns.boxplot(x=df[col], ax=axes[1])
    axes[1].set_title(f'Boxplot: {col}')

    sns.violinplot(x=df[col], ax=axes[2])
    axes[2].set_title(f'Violin: {col}')

    plt.tight_layout()
    plt.savefig(output_dir / f"dist_{col}.png")
    plt.show()

# ======================================================
# 7ï¸âƒ£ Trá»±c quan hÃ³a biáº¿n phÃ¢n loáº¡i
# ======================================================
for col in categorical_cols:
    vc = df[col].value_counts(dropna=False)
    n_unique = vc.shape[0]

    plt.figure(figsize=(8,4))
    sns.barplot(x=vc.index.astype(str), y=vc.values)
    plt.title(f'Biá»ƒu Ä‘á»“ cá»™t: {col}')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.savefig(output_dir / f"bar_{col}.png")
    plt.show()

    if n_unique <= 8:
        plt.figure(figsize=(6,6))
        plt.pie(vc.values, labels=[str(x) for x in vc.index], autopct='%1.1f%%', startangle=90)
        plt.title(f'Biá»ƒu Ä‘á»“ trÃ²n: {col}')
        plt.tight_layout()
        plt.savefig(output_dir / f"pie_{col}.png")
        plt.show()

# ======================================================
# 8ï¸âƒ£ Ma tráº­n tÆ°Æ¡ng quan vÃ  heatmap
# ======================================================
plt.figure(figsize=(10,8))
corr = df[numeric_cols].corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", center=0)
plt.title("Heatmap tÆ°Æ¡ng quan giá»¯a cÃ¡c biáº¿n sá»‘")
plt.tight_layout()
plt.savefig(output_dir / "correlation_heatmap.png")
plt.show()

# ======================================================
# 9ï¸âƒ£ Pairplot cho cÃ¡c biáº¿n sá»‘ (tá»‘i Ä‘a 6 cá»™t Ä‘áº§u)
# ======================================================
subset_cols = numeric_cols[:6]
sns.pairplot(df[subset_cols], diag_kind="hist", corner=True)
plt.suptitle("Pairplot (phÃ¢n tÃ­ch má»‘i quan há»‡ giá»¯a cÃ¡c biáº¿n sá»‘)", y=1.02)
plt.savefig(output_dir / "pairplot.png")
plt.show()

print("\nâœ… HoÃ n táº¥t EDA â€” cÃ¡c biá»ƒu Ä‘á»“ Ä‘Æ°á»£c lÆ°u trong thÆ° má»¥c /content/eda_outputs/")
```
