#Tạo ma trận khoảng cách và vẽ sơ đồ phân cấp (Dendrogram)

```python
# So sánh bốn phương pháp liên kết khác nhau
methods = ['ward', 'complete', 'average', 'single']
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle("Dendrogram với các phương pháp liên kết khác nhau", fontsize=16, fontweight='bold')

for i, method in enumerate(methods):
    row, col = divmod(i, 2)
    Z = linkage(X_scaled, method=method)
    dendrogram(Z, truncate_mode='lastp', p=30, ax=axes[row, col],
               leaf_font_size=9, show_contracted=True)
    axes[row, col].set_title(f"Liên kết: {method.capitalize()}")
    axes[row, col].set_xlabel("Mẫu / Nhóm gộp")
    axes[row, col].set_ylabel("Khoảng cách")

plt.tight_layout()
plt.show()
```

#Thực hiện phân cụm Agglomerative

```python
#Dựa vào dendrogram (cụ thể là phương pháp 'ward'), chọn n_clusters = 3 (tương ứng với 3 loài chim cánh cụt)
n_clusters = 3
agg_clustering = AgglomerativeClustering(n_clusters=n_clusters, linkage='ward')
labels_hc = agg_clustering.fit_predict(X_scaled)
```

```python
#Hiển thị các cụm sau khi phân cấp (Sử dụng 2 chiều đầu tiên để minh họa)
plt.figure(figsize=(8, 6))
plt.scatter(X_scaled[:,0], X_scaled[:,1], c=labels_hc, cmap='viridis', s=50)
plt.title(f"Agglomerative Clustering ({n_clusters} Cụm, Linkage: Ward)")
plt.xlabel("Chiều dài Mỏ (Đã chuẩn hóa)")
plt.ylabel("Chiều sâu Mỏ (Đã chuẩn hóa)")
plt.show()
```

#Đánh giá mô hình phân cụm

```python
print("\n--- Đánh giá mô hình phân cụm Agglomerative (k=3) ---")

#Silhouette Score
silhouette = silhouette_score(X_scaled, labels_hc)
print(f"Silhouette Score: {silhouette:.4f}")

#Davies-Bouldin Score
davies_bouldin = davies_bouldin_score(X_scaled, labels_hc)
print(f"Davies-Bouldin Score: {davies_bouldin:.4f}")

#Calinski-Harabasz Score
calinski_harabasz = calinski_harabasz_score(X_scaled, labels_hc)
print(f"Calinski-Harabasz Score: {calinski_harabasz:.4f}")
```
