```python
data = pd.read_csv("marketing_campaign.csv", encoding="latin1", sep="\t")
print("K√≠ch th∆∞·ªõc d·ªØ li·ªáu:", data.shape)
display(data.head())
```

```python
###AUTOVIZ#####
if 'data' in locals() and not data.empty: # Ch·ªâ ch·∫°y AutoViz n·∫øu d·ªØ li·ªáu ƒë√£ ƒë∆∞·ª£c n·∫°p th√†nh c√¥ng v√† kh√¥ng r·ªóng
    AV = AutoViz_Class()

    # T·∫°o th∆∞ m·ª•c l∆∞u bi·ªÉu ƒë·ªì n·∫øu ch∆∞a t·ªìn t·∫°i
    save_dir = "AutoViz_Plots"
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)

    report = AV.AutoViz(
        filename="",          # ƒê·ªÉ tr·ªëng v√¨ ƒë√£ truy·ªÅn dataframe qua dfte
        dfte=data,            # Truy·ªÅn dataframe ƒë√£ n·∫°p
        sep="\t",             # D·∫•u tab ph√¢n t√°ch d·ªØ li·ªáu
        depVar="",            # Kh√¥ng c√≥ bi·∫øn ph·ª• thu·ªôc c·ª• th·ªÉ
        header=0,
        verbose=2,            # Hi·ªÉn th·ªã chi ti·∫øt
        lowess=False,
        chart_format="html",
        max_rows_analyzed=len(data), # Ph√¢n t√≠ch to√†n b·ªô d·ªØ li·ªáu
        max_cols_analyzed=len(data.columns), # Ph√¢n t√≠ch to√†n b·ªô c·ªôt
        save_plot_dir=save_dir # L∆∞u bi·ªÉu ƒë·ªì v√†o th∆∞ m·ª•c ƒë√£ t·∫°o
    )

    print("\n‚úÖ B√°o c√°o AutoViz ƒë√£ ƒë∆∞·ª£c t·∫°o th√†nh c√¥ng!")
```

```python
#L∆ØU V√Ä T·∫¢I V·ªÄ B√ÅO C√ÅO + H√åNH
# N√©n th∆∞ m·ª•c ch·ª©a bi·ªÉu ƒë·ªì sau khi AutoViz ƒë√£ ch·∫°y xong
save_dir = "AutoViz_Plots" # Define save_dir
try:
    shutil.make_archive("AutoViz_Report", 'zip', save_dir)
    print(f"\n File AutoViz_Report.zip ƒë√£ s·∫µn s√†ng ƒë·ªÉ t·∫£i v·ªÅ t·ª´ th∆∞ m·ª•c '{save_dir}'")
    files.download("AutoViz_Report.zip")
except NameError:
    print("AutoViz_Plots directory not found. Please run AutoViz first.")
except Exception as e:
    print(f"An error occurred: {e}")
```
