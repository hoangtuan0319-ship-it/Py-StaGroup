```python
data = pd.read_csv("marketing_campaign.csv", encoding="latin1", sep="\t")
print("Kích thước dữ liệu:", data.shape)
display(data.head())
```

```python
###AUTOVIZ#####
if 'data' in locals() and not data.empty: # Chỉ chạy AutoViz nếu dữ liệu đã được nạp thành công và không rỗng
    AV = AutoViz_Class()

    # Tạo thư mục lưu biểu đồ nếu chưa tồn tại
    save_dir = "AutoViz_Plots"
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)

    report = AV.AutoViz(
        filename="",          # Để trống vì đã truyền dataframe qua dfte
        dfte=data,            # Truyền dataframe đã nạp
        sep="\t",             # Dấu tab phân tách dữ liệu
        depVar="",            # Không có biến phụ thuộc cụ thể
        header=0,
        verbose=2,            # Hiển thị chi tiết
        lowess=False,
        chart_format="html",
        max_rows_analyzed=len(data), # Phân tích toàn bộ dữ liệu
        max_cols_analyzed=len(data.columns), # Phân tích toàn bộ cột
        save_plot_dir=save_dir # Lưu biểu đồ vào thư mục đã tạo
    )

    print("\n✅ Báo cáo AutoViz đã được tạo thành công!")
```

```python
#LƯU VÀ TẢI VỀ BÁO CÁO + HÌNH
# Nén thư mục chứa biểu đồ sau khi AutoViz đã chạy xong
save_dir = "AutoViz_Plots" # Define save_dir
try:
    shutil.make_archive("AutoViz_Report", 'zip', save_dir)
    print(f"\n File AutoViz_Report.zip đã sẵn sàng để tải về từ thư mục '{save_dir}'")
    files.download("AutoViz_Report.zip")
except NameError:
    print("AutoViz_Plots directory not found. Please run AutoViz first.")
except Exception as e:
    print(f"An error occurred: {e}")
```
